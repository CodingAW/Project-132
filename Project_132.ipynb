```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv("P132.csv")

starMass = df['Mass'].tolist()
starRadius = df["Radius"].tolist()
starGravity = df["Gravity"].tolist()

starMass.sort()
starRadius.sort()
starGravity.sort()

plt.plot(starRadius, starMass)
plt.title("Radius and Mass Of The Star")
plt.xlabel("Radius")
plt.ylabel("Mass")

plt.show()
```

```python
plt.plot(starMass, starGravity)
plt.title("Mass vs. Gravity Of The Star")
plt.xlabel("Mass")
plt.ylabel("Gravity")

plt.show()
```

```python
plt.scatter(starRadius, starMass)
plt.title("Radius vs. Mass Of The Star")
plt.xlabel("Radius")
plt.ylabel("Mass")

plt.show()
```
